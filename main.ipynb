```python
# Because the dataset is too large, we'll clean the .json 
# following https://www.kaggle.com/code/prathamsharma123/clean-raw-json-tweets-data

import pandas as pd
from pandas.io.json import json_normalize
import warnings
warnings.filterwarnings("ignore")

# Drop extra columns
raw_tweets = pd.read_json(r'farmers-protest-tweets-2021-03-5.json', lines=True)
cols = ['url', 'date', 'content', 'id', 'user', 'retweetCount']
raw_tweets = raw_tweets[cols]
raw_tweets.head(10)
```

```python
# Extract user info
user_id = []
username = []
for user in raw_tweets['user']:
    user_id.append(user['id'])
    username.append(user['username'])
raw_tweets['userId'] = user_id
raw_tweets['username'] = username

raw_tweets.head(10)
```

```python
# Convert to DataFrame and remove duplicates 

tweets = pd.DataFrame(raw_tweets[['url', 'date', 'content', 'id', 'retweetCount', 'userId', 'username']])
tweets.drop_duplicates(subset=['id'], inplace=True)
tweets.head(10)
```
