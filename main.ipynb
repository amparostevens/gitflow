## Previous steps

```python
# Because the dataset is too large, we'll clean the .json 
# following https://www.kaggle.com/code/prathamsharma123/clean-raw-json-tweets-data

import pandas as pd
from pandas.io.json import json_normalize
import warnings
warnings.filterwarnings("ignore")

# Drop extra columns
raw_tweets = pd.read_json(r'farmers-protest-tweets-2021-03-5.json', lines=True)
cols = ['url', 'date', 'content', 'id', 'user', 'retweetCount']
raw_tweets = raw_tweets[cols]
raw_tweets.head(10)
```

```python
# Extract user info
user_id = []
username = []
for user in raw_tweets['user']:
    user_id.append(user['id'])
    username.append(user['username'])
raw_tweets['userId'] = user_id
raw_tweets['username'] = username

raw_tweets.head(10)
```

```python
# Convert to DataFrame and remove duplicates 

tweets = pd.DataFrame(raw_tweets[['url', 'date', 'content', 'id', 'retweetCount', 'userId', 'username']])
tweets.drop_duplicates(subset=['id'], inplace=True)
tweets.head(10)
```

## Fuctions

```python
def top_retweeted(tweets):
    top = tweets.sort_values(by='retweetCount', ascending=False)[['url', 'date', 'content', 'retweetCount']].head(10)
    return top
```

```python

def top_users(tweets):
    top = pd.DataFrame(tweets['username'].value_counts().head(10))
    sort = top.rename(columns = {'username': 'frequency'}, inplace = False)
    return sort
```

```python
def top_days(tweets):
    dates =pd.to_datetime(tweets['date'],format='%Y%m%d').dt.date
    top = pd.DataFrame(dates.value_counts().head(10))
    sort = top.rename(columns = {'date': 'frequency'}, inplace = False)
    return sort
```

```python
import re
hashs = tweets['content'].str.findall(r'#.*?(?=\s|$)').tolist()
top = pd.DataFrame([item for sublist in hashs for item in sublist]).value_counts().head(10)
top
```

## Main

```python
from IPython.display import display

# Define main function
def main(argument):
    if argument == 'top_retweeted':
        display(top_retweeted(tweets))
    elif argument == 'top_users':
        display(top_users(tweets))
    elif argument == 'top_days':
        display(top_days(tweets))
    # elif argument == 'top_hashtags':
    #     top_hashtags(tweets)
```
